### Feature Engineering
##### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
Label_Encoder=LabelEncoder()
```

```python
Salary_Data['Gender_Encode']=Label_Encoder.fit_transform(Salary_Data['Gender'])
Salary_Data['Degree_Encode']=Label_Encoder.fit_transform(Salary_Data['Degree'])
Salary_Data['Job_Title_Encode']=Label_Encoder.fit_transform(Salary_Data['Job_Title'])
```

### Data after label encoding

```python
Salary_Data.head()
```

```python
from sklearn.preprocessing import StandardScaler
std_Scaler=StandardScaler()
```

-#### Feature Scalling

```python
Salary_Data['Age_Scaler']=std_Scaler.fit_transform(Salary_Data[['Age']])
Salary_Data['Experience_years_Scaler']=std_Scaler.fit_transform(Salary_Data[['Experience_years']])
```

#### Data after scalling

```python
Salary_Data.head()
```

```python
Salary_Data.isna().sum()
```

### Dependent and independent variables

```python
x=Salary_Data[['Age_Scaler','Gender_Encode','Degree_Encode','Job_Title_Encode','Experience_years_Scaler']]
y=Salary_Data['Salary']
```

```python
x.head()
```

### splitting the data Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,y_train.shape  ## 80%
```

```python
x_test.shape ,y_test.shape ##20%
```

### Model development

```python
from sklearn.linear_model import LinearRegression
Linear_Regression_model=LinearRegression()
```

#### Model Training

```python
Linear_Regression_model.fit(x_train,y_train)
```

### Model predictions

```python
y_pred_lr=Linear_Regression_model.predict(x_test)
y_pred_lr
```

```python
df=pd.DataFrame({'y_Actual':y_test,'y_Predicted':y_pred_lr})
df
```

```python
df=pd.DataFrame({'y_Actual':y_test,'y_Predicted':y_pred_lr})
df['Error']=df['y_Actual']-df['y_Predicted']
df['abs_error']=abs(df['Error'])
df
```

```python
Mean_absolute_error=df['abs_error'].mean()
Mean_absolute_error
```

##### Model Evaluation

```python
from sklearn.metrics import accuracy_score,r2_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
print("r2_score :",r2_score(y_test,y_pred_lr)*100 )
print("mean_absolute_error:",mean_absolute_error(y_test,y_pred_lr))
print("mean_squared_error:",mean_squared_error(y_test,y_pred_lr))
print("Root mean_squared_error:",mean_squared_error(y_test,y_pred_lr)**(0.5))
```
