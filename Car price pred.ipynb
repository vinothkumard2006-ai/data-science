# Feature Engineering

```python
## One-hot encoding (normal data)

Car_data=pd.get_dummies(Car_data,columns=['Brand','Fuel_Type','Transmission','Model'],drop_first=True)
Car_data.head()

```

```python
## Labelencoding (ordinal data)

condition_map={'Used':0,'Like New':1,'New':2}
Car_data['Condition']=Car_data['Condition'].map(condition_map)
Car_data.tail()
```

```python
Car_data['Car_age']=2026-Car_data['Year']
Car_data.drop('Year',axis=1,inplace=True)
```

```python
Car_data.head()
```

```python
## Mileage Transformation (Skewed data)

Car_data['Mileage_log']=np.log1p(Car_data['Mileage'])
Car_data.drop('Mileage',axis=1,inplace=True)
```

```python
Car_data.head()
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
num_cols=['Engine_Size','Mileage_log','Car_age']
Car_data[num_cols]=scaler.fit_transform(Car_data[num_cols])
```

```python
Car_data.tail()
```

# Independent and dependent variable

```python
x=Car_data.drop('Price',axis=1)
y=Car_data['Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42) 
```

```python
from sklearn.linear_model import LinearRegression
model_lr=LinearRegression()
model_lr.fit(x_train,y_train)
```

```python
y_pred=model_lr.predict(x_test)
```

```python
## Model evaluation

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

print("R2 score:",r2_score(y_test,y_pred))
print("MAE:",mean_absolute_error(y_test,y_pred))
print("RMSE:",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
## not good model try another model
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rf=RandomForestRegressor(n_estimators=200,random_state=42)
model_rf.fit(x_train,y_train)
```

```python
y_pred_rf=model_rf.predict(x_test)
```

```python
print("R2 score:",r2_score(y_test,y_pred_rf))
print("MAE:",mean_absolute_error(y_test,y_pred_rf))
print("RMSE:",np.sqrt(mean_squared_error(y_test,y_pred_rf)))
```

```python
baseline_pred=np.full_like(y_test,y_train.mean())
print("Baseline R2:",r2_score(y_test,baseline_pred))
```
