```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,classification_report
```

```python
sns.get_dataset_names()
```

```python
MPG=sns.load_dataset("mpg")
MPG.head()
```

```python
MPG["cylinders"].unique()
```

```python
MPG["origin"].unique()
```

```python
MPG.drop('name',axis=1,inplace=True)
```

```python
MPG.tail()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
MPG['origin']=le.fit_transform(MPG['origin'])
```

```python
MPG.head()
```

```python
MPG.tail()
```

```python
MPG.isna().sum()
```

```python
MPG['horsepower'].unique()
```

```python
from scipy.stats import shapiro

stat,p=shapiro(MPG['horsepower'].dropna())
print("p.value:",p)
```

```python
MPG['horsepower'].fillna(MPG['horsepower'].mean(),inplace=True)
```

```python
MPG.isna().sum()
```

```python
MPG.describe().T
```

```python
MPG.drop_duplicates()
```

```python
MPG.shape
```

```python
MPG.dtypes
```

```python
x=MPG.drop(labels="cylinders",axis=1)
x
```

```python
y=MPG["cylinders"]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,shuffle=True,random_state=10)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

# model building

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_train_pred=model.predict(x_train)
```

```python
y_test_pred=model.predict(x_test)
```

```python
accuracy_score(y_train,y_train_pred)*100
```

```python
accuracy_score(y_test,y_test_pred)*100
```

```python
confusion_matrix(y_test,y_test_pred)
```

```python
print(classification_report(y_test,y_test_pred))
```
